#### Part 2-1 . üîç ÌîºÌï¥Í∞Ä Í∞ÄÏû• ÌÅ∞ ÏßÄÏó≠ÏùÄ Ïñ¥ÎîîÏùºÍπåÏöî?
> Í∞ÄÏû• ÌÅ∞ ÌîºÌï¥Î•º ÏûÖÏùÄ ÏßÄÏó≠Ïùò Ïû•ÎπÑ ÌååÍ¥¥ Í∞úÏàòÎ•º ÏÇ¥Ìé¥Î≥¥Î©¥ Ïïå Ïàò ÏûàÏùÑ  Í≤É Í∞ôÏäµÎãàÎã§.
Ï†ÑÌà¨Í∞Ä Í≤©Î†¨ÌïúÎßåÌÅº, ÌîºÌï¥ Î∞è Î∂ÄÏÉÅÏóê ÎåÄÌïú Í∞úÏàòÍ∞Ä ÌôïÏó∞ÌïòÍ≤å ÎßéÏù¥ ÎìúÎü¨ÎÇ† Í≤É ÏûÖÎãàÎã§.

````python
# Í∞ÄÏû• ÎßéÏùÄ ÌîºÌï¥Î•º ÏûÖÏùÄ ÎèÑÏãú Î∞è Í∞ÄÏû• ÌîºÌï¥Í∞Ä Ï†ÅÏùÄ ÏßÄÏó≠ÏùÑ Ï∞æÏïÑÎ¥ÖÏãúÎã§.
# ÌååÍ¥¥Îêú Ïû•ÎπÑÏùò Ï¥ù Í∞úÏàòÍ∞Ä Í∞ÄÏû• ÎßéÏùÄ ÏßÄÏó≠ Î∞è Í∞ÄÏû• Ï†ÅÏùÄ ÏßÄÏó≠ÏùÑ Ï∞æÏïÑÎ¥ÖÏãúÎã§.
# ÏÉàÎ°úÏö¥ Ïª¨ÎüºÏùÑ Ï∂îÍ∞ÄÌï©ÎãàÎã§.  - ÌååÍ¥¥Îêú Ïû•ÎπÑÏùò Ï¥ù Í∞úÏàò
eq_df['ÌååÍ¥¥Îêú_Ïû•ÎπÑÏùò_Ï¥ù_Í∞úÏàò'] = eq_df[eq_df.columns.difference(['ÎÇ†Ïßú','Í∞úÏ†ÑÏùº','ÌîºÌï¥Í∞Ä_ÌÅ∞_ÏßÄÏó≠'])].sum(axis=1)


#ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±Ìï©ÎãàÎã§.
region_df = eq_df[['ÌååÍ¥¥Îêú_Ïû•ÎπÑÏùò_Ï¥ù_Í∞úÏàò','ÌîºÌï¥Í∞Ä_ÌÅ∞_ÏßÄÏó≠']].sort_values(by='ÌååÍ¥¥Îêú_Ïû•ÎπÑÏùò_Ï¥ù_Í∞úÏàò', ascending = False)


# groubyÎ•º ÌÜµÌïòÏó¨ ÎèÑÏãúÎ≥Ñ ÌèâÍ∑†Í∞í Íµ¨ÌïòÍ∏∞
region_df.groupby('ÌîºÌï¥Í∞Ä_ÌÅ∞_ÏßÄÏó≠').mean().sort_values(by='ÌååÍ¥¥Îêú_Ïû•ÎπÑÏùò_Ï¥ù_Í∞úÏàò', ascending = False)
````

#### üó∫Í∞ÄÏû• ÎßéÏùÄ ÌîºÌï¥Î•º ÏûÖÏùÄ ÏßÄÏó≠Í≥º, Í∞ÄÏû• Ï†ÅÏùÄ ÌîºÌï¥Î•º ÏûÖÏùÄ ÏßÄÏó≠ÏùÑ ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.

Œ≤1: ÌöåÍ∑Ä Í≥ÑÏàò(slope, Í∏∞Ïö∏Í∏∞Î°ú Î≥º Ïàò ÏûàÎã§.) - ÏÑ§Î™ÖÎ≥ÄÏàò XÏùò Î≥ÄÌôîÏóê Îî∞Îùº Î∞òÏùëÎ≥ÄÏàò yÍ∞Ä Î∞òÏùëÌïòÎäî Ï†ïÎèÑ </br>

* ÌöåÍ∑Ä Í≥ÑÏàòÍ∞Ä ÏñëÏàòÎùºÎ©¥, xÍ∞íÏù¥ Ïª§ÏßÄÎ©¥ yÍ∞íÎèÑ Ï¶ùÍ∞ÄÌïòÎäî ÏÑ†Ïù¥ ÏÉùÍ∏¥Îã§.
* Í≥ÑÏàò ÏùåÏàòÏùº Îïå, xÍ∞íÏù¥ Ïª§ÏßÄÎ©¥ yÍ∞íÏùÄ Í∞êÏÜåÌïòÎäî ÏÑ†Ïù¥ ÏÉùÍ∏¥Îã§.

####ÏòàÏ∏° Ìï®ÏàòÎ•º ÏÉùÏÑ±ÌïòÍ≥†, ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Î•º Î∞òÎ≥µÌïòÏó¨ ÏòàÏ∏°ÌïòÍ≥†, coefficientÏùò ÏòÅÌñ•ÏùÑ ÏÑ§Î™ÖÌï¥Î¥ÖÏãúÎã§.

````python
def explain_prediction(column):
  y_pred = model.predict([[column]])
  pred = f'{int(column)} ÏÇ¨ÎßùÏûê Ïàò ÏòàÏ∏° :{int(y_pred[0])}Î™Ö, (1Í∞úÏùò Ïû•Í∞ëÏ∞®Îãπ Ï∂îÍ∞Ä ÏÇ¨ÎßùÏûê Ïàò {int(model.coef_[0])})' 
  return pred
````

````python
# Ïö∞Î¶¨Í∞Ä ÏòàÏ∏°ÌïòÍ≥†ÏûêÌïòÎäî Ïª¨ÎüºÏùÄ 'Ïû•Í∞ëÏ∞®'ÏûÖÎãàÎã§.
print(explain_prediction(620))
````

####ipywidgetsÎ•º ÏÇ¨Ïö©Ìï¥ÏÑú sqftÎ•º Î∞îÍæ∏Ïñ¥Í∞ÄÎ©∞ ÏòàÏ∏°ÌïòÏó¨ Î¥ÖÏãúÎã§.

````python
from ipywidgets import interact

# Îç∞ÏΩîÎ†àÏù¥ÌÑ∞ Ï∂îÍ∞Ä
# Îç∞ÏΩîÎ†àÏù¥ÌÑ∞ :Ìï®ÏàòÎ•º ÏàòÏ†ïÌïòÏßÄ ÏïäÎäî ÏÉÅÌÉúÏóêÏÑú Ï∂îÍ∞Ä Í∏∞Îä•ÏùÑ Íµ¨ÌòÑÌï† Îïå ÏÇ¨Ïö©.

@interact
def explain_prediction(column = (1, 1000)):
  y_pred = model.predict([[column]])
  pred = f'{int(column)} ÏÇ¨ÎßùÏûê Ïàò ÏòàÏ∏° :{int(y_pred[0])}Î™Ö, (1Í∞úÏùò Ïû•Í∞ëÏ∞®Îãπ Ï∂îÍ∞Ä ÏÇ¨ÎßùÏûê Ïàò {int(model.coef_[0])})' 
  return pred
````
